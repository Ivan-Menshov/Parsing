```python
import requests
import json
from pprint import pprint
import pandas as pd
```

```python
params = {
    'latitude': '55.710702',
    'longitude': '37.603747',
    'query': 'пицца',
    'cacheBreaker':'1663070371',
}
url = 'https://api.delivery-club.ru/api1.2/vendors/search'
```

```python
response = requests.get(url=url, params=params)

print(response.status_code)
pprint(response.json())
```

```python
resp = response.json()['vendors']
```

```python
shop_name,min_order_price,delivery_price,category_id = [],[],[],[]
for i in resp:
    shop_name.append(i['name'])
    min_order_price.append(i['delivery']['minOrderPrice']['value'])
    delivery_price.append(i['delivery']['price']['value'])
    category_id.append(i['categoryId'])
```

```python
data = {'shop_name': shop_name,
        'min_order_price': min_order_price,
        'delivery_price': delivery_price,
        'category_id': category_id}  
df = pd.DataFrame(data)  
print(df) 
```

```python
df_1 =df.loc[df['delivery_price']==0]
df_2 =df.loc[df['delivery_price']!=0]
m_1 = df_1['min_order_price'].mean()
m_2 = df_2['min_order_price'].mean()
```

```python
print(f'Доля с бесплатной доставкой = {df_1.shape[0] /df.shape[0] }')
print(f'Среднюю минимальную стоимость заказа (Бесплатной доставка) = {m_1}')
print(f'Доля с платной доставкой = {df_2.shape[0] /df.shape[0] }')
print(f'Среднюю минимальную стоимость заказа (Платная доставка) = {m_2}')
```

```python
dff_1=df_1.groupby('category_id').agg({ "category_id": "count"})/df_1.shape[0] 
dff_1.rename(columns={"category_id": "Доля по категориям"}, inplace=True)
print('Бесплатная доставка')
dff_1
```

```python
dff_2=df_2.groupby('category_id').agg({ "category_id": "count"})/df_2.shape[0]
dff_2.rename(columns={"category_id": "Доля по категориям"}, inplace=True)
print('Платная доставка')
dff_2
```
